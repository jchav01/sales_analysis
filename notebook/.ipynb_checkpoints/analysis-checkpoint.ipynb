## Data Extraction (SQL)

Data is extracted from PostgreSQL using SQL queries.
The queries include joins, aggregations, and window functions.


```python
from dotenv import load_dotenv
import os

load_dotenv()  # charge le fichier .env

# puis connexion
from sqlalchemy.engine import URL
from sqlalchemy import create_engine

url = URL.create(
    "postgresql+psycopg2",
    username="postgres",
    password=os.getenv("DB_PASSWORD"),
    host="127.0.0.1",
    port=5432,
    database="ecommerce",
)

engine = create_engine(url)

import pandas as pd
pd.read_sql("SELECT 1;", engine)
```

## Revenue Analysis

This section explores overall revenue and its evolution over time.


```python
import pandas as pd

query_monthly = """
SELECT
  DATE_TRUNC('month', o.order_purchase_timestamp) AS month,
  SUM(oi.price) AS revenue
FROM orders o
JOIN order_items oi ON oi.order_id = o.order_id
GROUP BY 1
ORDER BY 1;
"""

df_monthly = pd.read_sql(query_monthly, engine)
df_monthly.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df_monthly["month"], df_monthly["revenue"])
plt.title("Monthly Revenue")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
query_cat = """
SELECT
  COALESCE(ct.product_category_name_english, p.product_category_name) AS category,
  SUM(oi.price) AS revenue
FROM order_items oi
JOIN products p ON p.product_id = oi.product_id
LEFT JOIN category_translation ct
  ON ct.product_category_name = p.product_category_name
GROUP BY 1
ORDER BY revenue DESC
LIMIT 10;
"""

df_cat = pd.read_sql(query_cat, engine)
df_cat
```

```python
plt.figure(figsize=(8,5))
plt.barh(df_cat["category"], df_cat["revenue"])
plt.gca().invert_yaxis()
plt.title("Top 10 Categories by Revenue")
plt.xlabel("Revenue")
plt.tight_layout()
plt.show()
```

```python
query_aov = """
SELECT
  o.order_id,
  SUM(oi.price) AS order_value
FROM orders o
JOIN order_items oi ON oi.order_id = o.order_id
GROUP BY o.order_id;
"""
df_aov = pd.read_sql(query_aov, engine)
df_aov["order_value"].describe()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df_aov["order_value"], bins=50)
plt.title("Order Value Distribution")
plt.xlabel("Order Value")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

df_aov["order_value"].quantile([0.5, 0.75, 0.9, 0.95, 0.99])
```

```python
query_ship = """
SELECT
  o.order_id,
  SUM(oi.price) AS order_value,
  SUM(oi.freight_value) AS freight_value
FROM orders o
JOIN order_items oi ON oi.order_id = o.order_id
GROUP BY o.order_id;
"""
df_ship = pd.read_sql(query_ship, engine)

df_ship[["order_value", "freight_value"]].corr()
```

```python
plt.figure(figsize=(6,5))
plt.scatter(df_ship["order_value"], df_ship["freight_value"], s=5)
plt.title("Order Value vs Freight")
plt.xlabel("Order Value")
plt.ylabel("Freight Value")
plt.tight_layout()
plt.show()
```

```python
query_delivery = """
SELECT
  (order_delivered_customer_date - order_purchase_timestamp) AS delivery_time
FROM orders
WHERE order_delivered_customer_date IS NOT NULL
  AND order_purchase_timestamp IS NOT NULL;
"""
df_deliv = pd.read_sql(query_delivery, engine)

df_deliv["delivery_time"].describe()
```

```python
df_deliv["delivery_days"] = df_deliv["delivery_time"].dt.total_seconds() / 86400
df_deliv["delivery_days"].describe()
df_deliv["delivery_days"].quantile([0.5, 0.9, 0.95, 0.99])
```

```python
plt.figure(figsize=(8,5))
plt.hist(df_deliv["delivery_days"], bins=50)
plt.title("Delivery Time Distribution (days)")
plt.xlabel("Days")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
query_reviews = """
WITH order_value AS (
  SELECT order_id, SUM(price) AS order_value
  FROM order_items
  GROUP BY order_id
),
review AS (
  SELECT order_id, review_score
  FROM order_reviews
)
SELECT
  ov.order_value,
  r.review_score
FROM order_value ov
JOIN review r ON r.order_id = ov.order_id
WHERE r.review_score IS NOT NULL;
"""
df_rev = pd.read_sql(query_reviews, engine)

df_rev["group"] = df_rev["review_score"].apply(
    lambda x: "positive" if x >= 4 else ("negative" if x <= 2 else None)
)

df_rev_clean = df_rev.dropna(subset=["group"])

df_rev_clean["group"].value_counts()
```

```python
from scipy.stats import ttest_ind

pos = df_rev_clean[df_rev_clean["group"]=="positive"]["order_value"]
neg = df_rev_clean[df_rev_clean["group"]=="negative"]["order_value"]

pos.mean(), neg.mean(), ttest_ind(pos, neg, equal_var=False)
```
